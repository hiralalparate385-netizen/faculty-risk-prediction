```python
import sqlite3
import pandas as pd
import numpy as np
import os

print("🚀 Generating clean synthetic dataset for EDA")

# =========================
# PATH CONFIGURATION
# =========================
BASE_DIR = os.getcwd()
DB_PATH = os.path.join(BASE_DIR, "data", "faculty.db")

# =========================
# DATA CONFIGURATION
# =========================
NUM_RECORDS = 120
np.random.seed(42)

TERMS = [
    "Independence Term",
    "Festival Term",
    "Republic Term",
    "Colors Term"
]

EXAMS = [
    "Unit Test 1",
    "Mid Term",
    "Unit Test 2",
    "End Term"
]

records = []

# =========================
# DATA GENERATION (CLEAN)
# =========================
for faculty_id in range(1, NUM_RECORDS + 1):

    courses_assigned = np.random.choice(
        [1, 2, 3, 4],
        p=[0.2, 0.4, 0.3, 0.1]
    )

    weekly_teaching_hours = courses_assigned * np.random.randint(3, 6)
    preparation_hours_per_week = weekly_teaching_hours

    total_students_handled = courses_assigned * np.random.randint(30, 60)
    admin_roles_count = np.random.randint(0, 3)
    years_of_experience = np.random.randint(1, 25)

    term = np.random.choice(TERMS)
    exam_type = np.random.choice(EXAMS)

    # Workload scoring (clean & logical)
    workload_score = (
        weekly_teaching_hours +
        preparation_hours_per_week +
        admin_roles_count * 5 +
        (total_students_handled // 40)
    )

    if exam_type in ["Mid Term", "End Term"]:
        workload_score += 8

    if years_of_experience > 15:
        workload_score -= 4

    workload_risk = 1 if workload_score >= 45 else 0

    records.append([
        faculty_id,
        courses_assigned,
        weekly_teaching_hours,
        total_students_handled,
        admin_roles_count,
        years_of_experience,
        preparation_hours_per_week,
        term,
        exam_type,
        workload_risk
    ])

# =========================
# CREATE DATAFRAME
# =========================
df = pd.DataFrame(records, columns=[
    "faculty_id",
    "courses_assigned",
    "weekly_teaching_hours",
    "total_students_handled",
    "admin_roles_count",
    "years_of_experience",
    "preparation_hours_per_week",
    "term",
    "exam_type",
    "workload_risk"
])

# =========================
# STORE IN SQL
# =========================
os.makedirs(os.path.join(BASE_DIR, "data"), exist_ok=True)

conn = sqlite3.connect(DB_PATH)
df.to_sql(
    "faculty_workload",
    conn,
    if_exists="replace",
    index=False
)
conn.close()

print("✅ Dataset generated successfully")
print(f"📍 Database path: {DB_PATH}")
print(f"📊 Dataset shape: {df.shape}")
print("\n📌 Sample rows:")
print(df.head())
```
